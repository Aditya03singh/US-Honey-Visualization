**1) State with highest stocks**

```python
df_state = df.groupby("state")["stocks"].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="state",y="stocks",data=df_state)
plt.xlabel("States")
plt.ylabel("Stocks")
plt.xticks(rotation=90)
plt.title("States with highest stocks")
plt.show()
```

From the above graph we can conclude that the states with the highest stocks are:

1) South Dakota

2) North Dakota

3) California

4) Montana


and the states with the least stocks are ⁉

1) Oklahoma

2) South Carolina

3) Maryland

**2) State with highest colonies number**

```python
df_colonies = df.groupby("state")["colonies_number"].sum().sort_values(ascending=True).reset_index()
df_colonies
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="state",y="colonies_number",data=df_colonies)
plt.xlabel("States")
plt.ylabel("Colonies")
plt.xticks(rotation=90)
plt.title("States with highest Colonies")
plt.show()
```

From the above graph we can visualize that the states with the highest colonies number are **California** and **North Dakota**

**3) State with highest production**

```python
df_production = df.groupby("state")["production"].sum().sort_values().reset_index()
df_production
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="state",y="production",data=df_production)
plt.xlabel("States")
plt.ylabel("Honey Production")
plt.xticks(rotation=90)
plt.title("States with highest Production")
plt.show()
```

**4) State with highest yield per colony**

```python
df_yield = df.groupby("state")["yield_per_colony"].sum().sort_values().reset_index()
df_yield
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="state",y="yield_per_colony",data=df_yield)
plt.xlabel("States")
plt.ylabel("Yields Per Colony")
plt.xticks(rotation=90)
plt.title("States with highest Yield per Colony")
plt.show()
```

**5) Year with highest mean average price**

```python
df_year = df.groupby("year")["average_price"].mean().sort_values().reset_index()
df_year
```

```python
plt.figure(figsize=(20,10))
ax=sns.barplot(x="year",y="average_price",data=df_year,order = df_year.sort_values('average_price').year,errwidth =  0)
ax.bar_label(ax.containers[0])
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Year with highest mean average price")
plt.show()
```

```python
plt.figure(figsize=(20,10))
ax=sns.scatterplot(x="year",y="average_price",data=df_year)
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Year with highest mean average price")
plt.show()
```

**6) Year with highest mean production**

```python
df_year_production = df.groupby("year")["production"].mean().sort_values().reset_index()
df_year_production
```

```python
plt.figure(figsize=(25,5))
ax=sns.barplot(x="year",y="production",data=df_year_production,order = df_year_production.sort_values('production').year,errwidth =  0)
ax.bar_label(ax.containers[0])
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Year with highest mean Production")
plt.show()
```

**7) Year with cheapest mean average price**

```python
df_yr = df.groupby("year")["average_price"].min().sort_values().reset_index()
df_yr
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="year",y="average_price",data= df_yr)
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Year with cheapest mean average price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="year",y="average_price",data= df_yr)
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Year with cheapest mean average price")
plt.xticks(rotation=90)
plt.show()
```

**8) Year with highest average price**

```python
df_yr_avg = df.groupby("year")["average_price"].max().sort_values().reset_index()
df_yr_avg
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x="year",y="average_price",data= df_yr_avg,order=df_yr_avg.sort_values("average_price").year)
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Year with Highest average price")
plt.xticks(rotation=90)
plt.show()
```
